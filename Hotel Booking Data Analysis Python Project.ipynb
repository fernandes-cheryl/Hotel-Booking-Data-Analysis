# Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
# value_counts returns the category name and number of times it is present in the colummn
# normalize turns the number into percentage
```

```python
# Visualizing canceled and not canceled
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
plt.title('Reservation status in different hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
# Filtering by resort hotel and calculating percentage of canceled reservations
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Filtering by city hotel and calculating percentage of canceled reservations
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Grouping based on reservation_status_date and taking the mean of adr for that date
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Visualizing 
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel')
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend()
plt.show()
```

```python
# This visualization shows that city hotel costs lesser than resort hotels
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('Reservation Status Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend('Not canceled', 'Canceled')
plt.show()
```

```python
# Effect of price on cancellation 
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
#From this visualization it shows that higher price results in more cancelations
```

```python
# Cancelations based on top-10 countries
```

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
# [:10] for descending order
plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with reservations canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)
#in increasing order canceled adr

not_canceled_data =df[df['is_canceled'] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', fontsize = 25)
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='canceled')
plt.legend()
```

```python
# The data is inconsistent so we are only taking data from 2016-01 to 2017-09 
```
